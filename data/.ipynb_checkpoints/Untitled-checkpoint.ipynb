```python
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.shape
```

```python
booleans = []
for length in movies.duration:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[0:5]
```

```python
is_long = pd.Series(booleans)
```

```python
is_long.head()
```

```python
movies['genre'].head()
```

```python
movies[is_long].head()
```

```python
is_long = movies.duration >= 200 #make series for putting condition in them
```

```python
movies[is_long].head()
```

```python
movies[movies.duration >= 200].head() #
```

```python
movies[movies.duration >= 200]['genre'].head()
```

```python
movies.loc[movies.duration >= 200, 'genre']#prctical way to do condition and get a column at the same time
```

```python
movies.loc[(movies.duration >= 200) & (movies.genre == 'Drama')]
```

```python
movies[(movies.duration >= 200) & (movies.genre == 'Drama')]
```

```python
movies.loc[(movies.duration >= 200) & (movies.genre == 'Drama')]
```

```python
movies.loc[(movies.duration >= 200) & (movies.genre.isin(['Drama' , 'Action']))]
```
